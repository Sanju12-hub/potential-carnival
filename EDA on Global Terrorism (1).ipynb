```python
terr2=terror.copy()
terr2.head()                    #copy the dataset into new dataframe
```

```python
terr2.dropna(axis=1,how='all')         #Dropping null values
```

# Rename the columns for better understand

```python
terr2.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'State','region_txt':'Region','target1':'Target','nkill':'Killed','nwound':'Wounded','gname':'Group','weaptype1_txt':'Weapon_type','targtype1_txt':'Target_type','attacktype1_txt':'AttackType','summary':'Summary','lattitude':'Lattitude','longitude':'Longitude','motive':'Motive'},inplace=True)
```

```python
terr2=terr2[['Year','Month','Day','Country','State','Region','Longitude','AttackType','Wounded','Target','Summary','Group',
            'Target_type','Weapon_type','Motive']]
terr2['Wounded']=terr2['Wounded'].fillna(0).astype(int)
```

```python
terr2['casuals']=terr2['Wounded']
terr2.head()
```

```python
terr2.isnull().sum()
```

```python
print('The country with highest terror attack:',terr2['Country'].value_counts().index[0])
```

# Correlation

```python
a = terr2.corr()
mask = np.triu(np.ones_like(a, dtype=bool))
c, ax = plt.subplots(figsize=(18,12))
cmap = sns.diverging_palette(250,20,as_cmap=True)
sns.heatmap(a, mask=mask, cmap=cmap, vmax=.3,center=0,square=True,linewidth=.5, cbar_kws={'shrink': .5})
plt.title("Corrilation matrix (data)")
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(a,annot=True,annot_kws={'size':10},linewidth=1,vmin=0,vmax=0.5,cmap='ocean_r')
plt.title('correlayion matrix (data)')
plt.show()
```

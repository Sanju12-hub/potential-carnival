```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import lux             #data visualization library
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
path = 'https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv'
```

```python
s_data = pd.read_csv(path)
```

```python
s_data.head()
```

# preparing the data

```python
s_data.columns
```

```python
s_data.shape
```

```python
train,test = train_test_split(s_data,test_size=0.25,random_state=123)
```

```python
train.shape
```

```python
test.shape
```

```python
train_x=train.drop('Scores',axis=1)
test_y=train["Scores"]
```

```python
test_x=test.drop("Scores",axis=1)
test_y=train["Scores"]
```

# training the algorithm

```python
lr=LinearRegression()
```

```python
lr.fit(train_x,train_y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
#plotting the regression line
line = lr.coef_*train_x+lr.intercept_
plt.scatter(train_x,train_y)
plt.plot(train_x,line);
plt.show()
```

```python
pr=lr.predict(test_x)
```

# making prediction

```python
list(zip(test_y,pr))
```

# evaluating the model

```python
from sklearn.metrics import mean_squared_error   #RMSE
```

# solution

```python
hour = [9.25]
own_pr=lr.predict([hour])
print('no of hours = {}'.format([hour]))
print("prediction score={}".format(own_pr[0]))
```
